```python
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
```

```python
# read csv file
file_path2 = "1976-2018-house2.csv"

# create the data frame 
house_df = pd.read_csv(file_path2, encoding='iso-8859-1')
house_df.head()
```

```python
# dropping # of years we do not need
house_df = house_df[house_df.year != 1976]
house_df = house_df[house_df.year != 1978]
house_df = house_df[house_df.year != 1982]
house_df = house_df[house_df.year != 1980]
house_df = house_df[house_df.year != 1984]
house_df = house_df[house_df.year != 1986]
house_df = house_df[house_df.year != 1988]
house_df = house_df[house_df.year != 1990]
house_df = house_df[house_df.year != 1994]
house_df = house_df[house_df.year != 1992]
house_df = house_df[house_df.year != 1996]
house_df = house_df[house_df.year != 2000]
house_df = house_df[house_df.year != 1998]
house_df = house_df[house_df.year != 2002]
house_df = house_df[house_df.year != 2004]
house_df = house_df[house_df.year != 2006]
house_df = house_df[house_df.year != 2008]
house_df = house_df[house_df.year != 2010]
house_df = house_df[house_df.year != 2012]
house_df = house_df[house_df.year != 2014]
house_df = house_df[house_df.year != 2018]
house_df.head()
```

```python
# Drop the columns that we don't need
house_df = house_df.drop(columns=["year", "state_fips", "state_cen", "state_ic", "office", "stage", "runoff",
                                  "mode", "unofficial", "version"])
house_df.head()
```

```python
#renaming columns
house_df = house_df.rename(columns={"state": "State", "state_po": "State Abbreviation", "district": "District Number",
                                    "special": "Special Election", "candidate": "Candidate", "party": "Party",
                                    "candidatevotes": "Candidate Votes", "totalvotes": "Total Votes"})
house_df.head()
```

```python
# Create a new column with the % of the vote that the candidate received
house_df["Candidate Percent"] = house_df["Candidate Votes"] / house_df["Total Votes"]
house_df.head()
```

```python
# change NaN values to 0
house_df.fillna(0, inplace=True)
house_df.head()
```

```python
# Export the cleaned data tocsv
house_df.to_csv("Clean Data/house_overall_2016_clean.csv", index=False)
```
