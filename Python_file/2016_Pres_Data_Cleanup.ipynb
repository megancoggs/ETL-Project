```python
# import dependencies
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
```

```python
# read csv file
file_path = "countypres_2000-2016.csv"

# create the data frame 
pres_df = pd.read_csv(file_path)
pres_df.head()
```

```python
#pres_df.columns=['year','state','state_po','county','office','candidate','party','candidatevotes','totalvotes']
#pres_df.head()
pres_df = pres_df.drop(columns=['FIPS', 'version'])
pres_df.head()
```

```python
#dropping # of years not needed

pres_df = pres_df[pres_df.year != 2000]
pres_df = pres_df[pres_df.year != 2004]
pres_df = pres_df[pres_df.year != 2008]
pres_df = pres_df[pres_df.year != 2012]
pres_df.head()
```

```python
#renaming columns
pres_df = pres_df.rename(columns={"state": "State", "state_po": "State Abbreviation", "county": "County",
                                    "office": "Office", "candidate": "Candidate", "party": "Party",
                                    "candidatevotes": "Candidate Votes", "totalvotes": "Total Votes"})
pres_df.head()
```

```python
# change NaN values to 0
pres_df.fillna(0, inplace=True)
pres_df.head()
```

```python
# Create a new column with the % of the vote that the candidate received
pres_df["Candidate Percent"] = pres_df["Candidate Votes"] / pres_df["Total Votes"]
pres_df.head()
```

```python
# Export the cleaned data tocsv
pres_df.to_csv("Clean Data/pres_overall_2016_clean.csv", index=False)
```
