```python
import pandas as pd
```

```python
race_2018_df = pd.read_excel('race_2018.xlsx', index_col=1)
```

```python
race_2018_df
```

```python
#get rid of first few rows
race_2018_df = race_2018_df.iloc[1:]
race_2018_df.head(10)
```

```python
#get rid of uninteresting columns
race_2018_df = race_2018_df.drop(race_2018_df.columns[[40,39,38,36,35,34,32,31,30,28,27,26,24,23,22,20,19,18,16,15,14,12,11,10,8,7,6,4,0]], axis=1)
race_2018_df.head(20)
```

```python
#rename columns

race_2018_df = race_2018_df.rename(columns={"Unnamed: 2" : "state", 
                                          "Unnamed: 3" : "district_number", 
                                          "Unnamed: 4" : "voting_age_population", 
                                          "Unnamed: 6" : "white",
                                          "Unnamed: 10" : "black",
                                          "Unnamed: 14" : "asian",
                                          "Unnamed: 18" : "american_indian_and_alaska_native",
                                          "Unnamed: 22" : "native_hawaiian_and_other_pacific_islander",
                                          "Unnamed: 26" : "some_other_race",
                                          "Unnamed: 30" : "two_or_more_races",
                                          "Unnamed: 34" : "hispanic",
                                          "Unnamed: 38" : "white_non_hispanic"})

race_2018_df
```

```python
#drop more rows using a different method for no reason..

race_2018_df = race_2018_df.drop(race_2018_df.index[[0,1,2]])
race_2018_df.head(20)
```

```python
#reset index to pop abbreviations out into own column

race_2018_df = race_2018_df.reset_index()
race_2018_df
```

```python
#rename index column

race_2018_df = race_2018_df.rename(columns={"index":"state_abbreviation"})
race_2018_df
```

```python
#Add Congressional District column to eventually use as PK, using a simpler method from before

race_2018_df['congressional_district'] = race_2018_df["state_abbreviation"].astype(str) + "-" + race_2018_df["district_number"].astype(str)
race_2018_df
```

```python
#Splendid! export to csv

race_2018_df.to_csv('race2018.csv', index=False, header=True)
```
