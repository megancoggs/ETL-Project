```python
import pandas as pd
```

```python
edu_2018_df = pd.read_excel('educational_attainment_2018.xlsx', index_col=1)
```

```python
edu_2018_df
```

```python
#get rid of first few rows
edu_2018_df = edu_2018_df.iloc[1:]
edu_2018_df.head(10)
```

```python
#get rid of uninteresting columns
edu_2018_df = edu_2018_df.drop(edu_2018_df.columns[[40,39,38,36,35,34,32,31,30,28,27,26,24,23,22,20,19,18,16,15,14,12,11,10,8,7,6,4,0]], axis=1)
edu_2018_df.head(10)
```

```python
#rename columns

edu_2018_df = edu_2018_df.rename(columns={"Unnamed: 2" : "State", 
                                          "Unnamed: 3" : "District Number", 
                                          "Unnamed: 4" : "Voting-Age Population", 
                                          "Unnamed: 6" : "Less than 9th Grade",
                                          "Unnamed: 10" : "9th to 12 Grade, No Diploma",
                                          "Unnamed: 14" : "High School Graduate",
                                          "Unnamed: 18" : "Some College, No Degree",
                                          "Unnamed: 22" : "Associate's Degree",
                                          "Unnamed: 26" : "Bachelor's Degree",
                                          "Unnamed: 30" : "Graduate Degree",
                                          "Unnamed: 34" : "High School or More",
                                          "Unnamed: 38" : "Bachelor's Degree or More"})

edu_2018_df
                                          
```

```python
#drop more rows using a different way for no reason..

edu_2018_df = edu_2018_df.drop(edu_2018_df.index[[0,1,2]])
edu_2018_df.head(20)
```

```python
#reset index to pop abbreviations out into own column

edu_2018_df = edu_2018_df.reset_index()
edu_2018_df
```

```python
edu_2018_df = edu_2018_df.rename(columns={"index":"State Abbreviation"})
edu_2018_df
```

```python
#Add Congressional District column to eventually use as PK

edu_2018_df['Congressional District'] = ['-'.join(i) for i in zip(edu_2018_df["State Abbreviation"].map(str),edu_2018_df["District Number"].map(str))]
edu_2018_df
```

```python
#Noice. export to csv

edu_2018_df.to_csv('edu2018.csv', index=False, header=True)
```
