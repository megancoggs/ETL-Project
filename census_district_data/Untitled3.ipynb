```python
import pandas as pd
```

```python
#import demo data

age_2018_df = pd.read_csv('age2018.csv', index_col=1)
edu_2018_df = pd.read_csv('edu2018.csv', index_col=1)
sap_2018_df = pd.read_csv('sap2018.csv', index_col=1)
race_2018_df = pd.read_csv('race2018.csv', index_col=1)
```

```python
#check importing

age_2018_df = age_2018_df.reset_index()
sap_2018_df = sap_2018_df.reset_index()
edu_2018_df = edu_2018_df.reset_index()
race_2018_df = race_2018_df.reset_index()
```

```python
#peek at age_2018_df
age_2018_df
```

```python
#combine data into one table, starting with the columns all tables share and then the cols unique to the age_2018 df

combined_2018_df = age_2018_df[["State", 
                                "State Abbreviation", 
                                "District Number", 
                                "Congressional District", 
                                "Voting-Age Population", 
                                "18-29 Years Old", 
                                "30-44 Years Old", 
                                "30-44 Years Old", 
                                "45-64 Years Old", 
                                "65+ Years Old"]]
                    
combined_2018_df
                   
```

```python
#add in new sex and poverty data from sap_2018_df

#sap_2018_df.head(10)

combined_2018_df = combined_2018_df.join(sap_2018_df[["Men", "Women", "Above Poverty Level", "Below Poverty Level"]])

combined_2018_df
```

```python
#add in educational attainment data from edu_2018_df

#edu_2018_df.head()

combined_2018_df = combined_2018_df.join(edu_2018_df[["Less than 9th Grade",
                                                      "9th to 12 Grade, No Diploma",
                                                      "High School Graduate",
                                                      "Some College, No Degree",
                                                      "Associate's Degree",
                                                      "Bachelor's Degree",
                                                      "Graduate Degree",
                                                      "High School or More",
                                                      "Bachelor's Degree or More"]])

combined_2018_df
```

```python
#Lastly, add in race data from race_2018_df

race_2018_df.head()

combined_2018_df = combined_2018_df.join(race_2018_df[["White",
                                                      "Black",
                                                      "Asian",
                                                      "American Indian and Alaska Native",
                                                      "Native Hawaiian and Other Pacific Islander",
                                                      "Some Other Race",
                                                      "Two or More Races",
                                                      "Hispanic",
                                                      "White Non-Hispanic"]])

combined_2018_df
```

```python
#Wonderful. Now export combined_2018_df as csv

combined_2018_df.to_csv('combined2018.csv', index=False, header=True)
```
