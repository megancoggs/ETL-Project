```python
import pandas as pd
```

```python
sex_and_poverty_2018_df = pd.read_excel('sex_and_poverty_2018.xlsx', index_col=1)
```

```python
sex_and_poverty_2018_df.head(10)
```

```python
#eliminate first couple of rows, rename df to something shorter
sap_2018_df = sex_and_poverty_2018_df.iloc[2:]
sap_2018_df.head(10)
```

```python
#get rid of uninteresting columns
sap_2018_df = sap_2018_df.drop(sap_2018_df.columns[[18,17,16,14,12,11,10,8,7,6,4,0]], axis=1)

sap_2018_df.head(20)
```

```python
#rename columns
sap_2018_df = sap_2018_df.rename(columns={"Unnamed: 2" : "State",
                                          "Unnamed: 3" : "District Number",
                                          "Unnamed: 4" : "Voting-Age Population",
                                          "Unnamed: 6" : "Men", "Unnamed: 10" : "Women",
                                          "Unnamed: 14" : "Above Poverty Level",
                                          "Unnamed: 16" : "Below Poverty Level"})
sap_2018_df
```

```python
#remove first two rows, and some at the end after..

sap_2018_df = sap_2018_df.iloc[2:]
#sap_2018_df = sap_2018_df.drop("NaN")
sap_2018_df.head(10)
```

```python
#figure out how many rows at the end need to be removed
sap_2018_df.tail(20)
```

```python
#remove last 7 rows
sap_2018_df = sap_2018_df.iloc[:-7]
sap_2018_df
```

```python
#reset index to pop the abbreviations into a manipulable column
sap_2018_df = sap_2018_df.reset_index()
sap_2018_df
```

```python
#rename "index" column
sap_2018_df = sap_2018_df.rename(columns={"index":"State Abbreviation"})
sap_2018_df
```

```python
#add congressional district column

sap_2018_df['Congressional District'] = ['-'.join(i) for i in zip(sap_2018_df["State Abbreviation"].map(str),sap_2018_df["District Number"].map(str))]

sap_2018_df
```

```python
#May as well split dataframe into a sex dataframe and a poverty level database

sex_2018_df = sap_2018_df[['State', 'District Number', 'Congressional District', 'Voting-Age Population', 'Men', 'Women']]

poverty_2018_df = sap_2018_df[['State', 'District Number', 'Congressional District', 'Voting-Age Population', 'Above Poverty Level', 'Below Poverty Level']]
```

```python
#Lovely - print each to csv file (or all three)
sap_2018_df.to_csv('sap2018.csv', index=False, header=True)
sex_2018_df.to_csv('sex2018.csv', index=False, header=True)
poverty_2018_df.to_csv('poverty2018.csv', index=False, header=True)
```
