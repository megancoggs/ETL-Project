```python
import pandas as pd
```

```python
#import demo data

age_2018_df = pd.read_csv('age2018.csv', index_col=1)
edu_2018_df = pd.read_csv('edu2018.csv', index_col=1)
sap_2018_df = pd.read_csv('sap2018.csv', index_col=1)
race_2018_df = pd.read_csv('race2018.csv', index_col=1)
```

```python
age_2018_df = age_2018_df.reset_index()
sap_2018_df = sap_2018_df.reset_index()
edu_2018_df = edu_2018_df.reset_index()
race_2018_df = race_2018_df.reset_index()
```

```python
#peek at age_2018_df
age_2018_df
```

```python
#combine data into one table, starting with the columns all tables share and then also the cols unique to the age_2018 df

combined_2018_df = age_2018_df[["state", 
                                "state_abbreviation", 
                                "district_number", 
                                "congressional_district", 
                                "voting_age_population", 
                                "age_18_29_years_old", 
                                "age_30_44_years_old", 
                                "age_45_64_years_old", 
                                "age_65_and_older"]]
                    
combined_2018_df
                   
```

```python
#add in new sex and poverty data from sap_2018_df

#sap_2018_df.head(10)

combined_2018_df = combined_2018_df.join(sap_2018_df[["men", "women", "above_poverty_level", "below_poverty_level"]])

combined_2018_df
```

```python
#add in educational attainment data from edu_2018_df

#edu_2018_df.head()

combined_2018_df = combined_2018_df.join(edu_2018_df[["less_than_9th_grade",
                                                      "high_school_no_diploma",
                                                      "high_school_graduate",
                                                      "some_college_no_degree",
                                                      "associates_degree",
                                                      "bachelors_degree",
                                                      "graduate_degree",
                                                      "high_school_or_more",
                                                      "bachelors_degree_or_more"]])

combined_2018_df
```

```python
#Lastly, add in race data from race_2018_df

race_2018_df.head()

combined_2018_df = combined_2018_df.join(race_2018_df[["white",
                                                      "black",
                                                      "asian",
                                                      "american_indian_and_alaska_native",
                                                      "native_hawaiian_and_other_pacific_islander",
                                                      "some_other_race",
                                                      "two_or_more_races",
                                                      "hispanic",
                                                      "white_non_hispanic"]])

combined_2018_df
```

```python
#Wonderful. Now export combined_2018_df as csv

combined_2018_df.to_csv('combined2018.csv', index=False, header=True)
```
