```python
import pandas as pd
```

```python
age_2018_df = pd.read_excel('age_2018.xlsx', index_col=1)
```

```python
#get rid of first few rows
age_2018_df = age_2018_df.iloc[1:]
age_2018_df.head(10)
```

```python
#get rid of uninteresting columns
age_2018_df = age_2018_df.drop(age_2018_df.columns[[20,19,18,16,15,14,12,11,10,8,7,6,4,0]], axis=1)


age_2018_df.head(20)
```

```python
#rename columns
age_2018_df = age_2018_df.rename(columns={"Unnamed: 2" : "State", 
                                          "Unnamed: 3" : "District Number", 
                                          "Unnamed: 4" : "Voting-Age Population", 
                                          "Unnamed: 6" : "18-29 Years Old", 
                                          "Unnamed: 10" : "30-44 Years Old", 
                                          "Unnamed: 14" : "45-64 Years Old", 
                                          "Unnamed: 18" : "65+ Years Old"})
age_2018_df.head(10)
```

```python
#df.drop(df.index[[1,3]])

age_2018_df = age_2018_df.drop(age_2018_df.index[[0,1,2]])
age_2018_df.head(20)
```

```python
age_2018_df = age_2018_df.reset_index()
age_2018_df
```

```python
age_2018_df = age_2018_df.rename(columns={"index":"State Abbreviation"})
age_2018_df
```

```python
#Add Congressional District column to eventually use as PK
#age_2018_df['Congressional District'] = age_2018_df[['State', 'District Number']].astype(str).sum(axis=1)

age_2018_df['Congressional District'] = ['-'.join(i) for i in zip(age_2018_df["State Abbreviation"].map(str),age_2018_df["District Number"].map(str))]
age_2018_df
```

```python
age_2018_df.to_csv('age2018.csv', index=False, header=True)
```
