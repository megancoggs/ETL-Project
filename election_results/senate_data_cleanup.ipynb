```python
# import dependencies
import numpy as np
import pandas as pd
```

```python
# read csv file
file_path = "data_original/1976-2018-senate.csv"

# create the data frame 
senate_df_original = pd.read_csv(file_path, encoding='iso-8859-1')
senate_df_original.head()
```

```python
# Make a copy
senate_df = senate_df_original
```

```python
# dropping years we do not need
senate_df = senate_df[senate_df.year != 1976]
senate_df = senate_df[senate_df.year != 1978]
senate_df = senate_df[senate_df.year != 1980]
senate_df = senate_df[senate_df.year != 1982]
senate_df = senate_df[senate_df.year != 1984]
senate_df = senate_df[senate_df.year != 1986]
senate_df = senate_df[senate_df.year != 1988]
senate_df = senate_df[senate_df.year != 1990]
senate_df = senate_df[senate_df.year != 1992]
senate_df = senate_df[senate_df.year != 1994]
senate_df = senate_df[senate_df.year != 1996]
senate_df = senate_df[senate_df.year != 1998]
senate_df = senate_df[senate_df.year != 2000]
senate_df = senate_df[senate_df.year != 2002]
senate_df = senate_df[senate_df.year != 2004]
senate_df = senate_df[senate_df.year != 2006]
senate_df = senate_df[senate_df.year != 2008]
senate_df = senate_df[senate_df.year != 2010]
senate_df = senate_df[senate_df.year != 2012]
senate_df = senate_df[senate_df.year != 2014]
senate_df.head()
```

```python
# dropping columns
senate_df = senate_df.drop(columns=["state_fips", "state_cen", "state_ic", "office", "stage", "writein",
                                  "mode", "unofficial", "version"])
senate_df.head()
```

```python
#renaming columns
senate_df = senate_df.rename(columns={"year": "Year", "state": "State", "state_po": "State Abbreviation", "district": "District Number",
                                    "special": "Special Election", "candidate": "Candidate", "party": "Party",
                                    "candidatevotes": "Candidate Votes", "totalvotes": "Total Votes"})
senate_df.head()
```

```python
# Create a new column with the % of the vote that the candidate received
senate_df["Candidate Percent"] = senate_df["Candidate Votes"] / senate_df["Total Votes"]
senate_df.head()
```

```python
# Change all parties other than democrat and republican to "other"
senate_df.loc[(senate_df["Party"] != "republican") & (senate_df["Party"] != "democrat"), "Party"] = "other"
senate_df.loc[senate_df["Party"] == "other"]
```

```python
# View final DataFrame
senate_df.head()
```

```python
# Export the cleaned data tocsv
senate_df.to_csv("data_clean/senate_clean.csv", index=False)
```
