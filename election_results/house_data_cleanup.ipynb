```python
import numpy as np
import pandas as pd
```

```python
# read csv file
file_path2 = "data_original/1976-2018-house2.csv"

# create the data frame 
house_df_original = pd.read_csv(file_path2, encoding='iso-8859-1')
house_df_original.head()
```

```python
# Make a copy
house_df = house_df_original

# Grab data for 2016 and 2018 only
# dropping # of years we do not need
house_df = house_df[house_df.year != 1976]
house_df = house_df[house_df.year != 1978]
house_df = house_df[house_df.year != 1982]
house_df = house_df[house_df.year != 1980]
house_df = house_df[house_df.year != 1984]
house_df = house_df[house_df.year != 1986]
house_df = house_df[house_df.year != 1988]
house_df = house_df[house_df.year != 1990]
house_df = house_df[house_df.year != 1994]
house_df = house_df[house_df.year != 1992]
house_df = house_df[house_df.year != 1996]
house_df = house_df[house_df.year != 2000]
house_df = house_df[house_df.year != 1998]
house_df = house_df[house_df.year != 2002]
house_df = house_df[house_df.year != 2004]
house_df = house_df[house_df.year != 2006]
house_df = house_df[house_df.year != 2008]
house_df = house_df[house_df.year != 2010]
house_df = house_df[house_df.year != 2012]
house_df = house_df[house_df.year != 2014]
house_df.head()
```

```python
# Drop the columns that we don't need
house_df = house_df.drop(columns=["state_fips", "state_cen", "state_ic", "office", "stage", "runoff", "writein",
                                  "mode", "unofficial", "version"])
house_df.head()
```

```python
#renaming columns
house_df = house_df.rename(columns={"state_po": "state_abbreviation", "district": "district_number",
                                    "special": "special_election", "candidatevotes": "candidate_votes", 
                                    "totalvotes": "total_votes"})
house_df.head()
```

```python
# Create new column with unique identifier for state-district
house_df["congressional_district"] = house_df["state_abbreviation"] + "-" + (house_df["district_number"]).astype("str")
house_df.head()
```

```python
# Create a new column with the % of the vote that the candidate received
house_df["candidate_percent"] = house_df["candidate_votes"] / house_df["total_votes"]
house_df.head()
```

```python
# Change all parties other than democrat and republican to "other"
house_df.loc[(house_df["party"] != "republican") & (house_df["party"] != "democrat"), "party"] = "other"
house_df.loc[house_df["party"] == "other"]
```

```python
# Reorder columns
house_df = house_df[["year", "congressional_district", "state", "state_abbreviation", "district_number", "special_election",
                                 "candidate", "party", "candidate_votes", "total_votes", "candidate_percent"]]
house_df.head()
```

```python
# Export the cleaned data tocsv
house_df.to_csv("data_clean/house_clean.csv", index=False)
```
