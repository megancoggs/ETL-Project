```python
import pandas as pd
```

```python
#importing data
unemploy_2016_df = pd.read_csv('state_unemployment_2016.csv', index_col=1)
unemploy_2018_df = pd.read_csv('state_unemployment_2018.csv', index_col=1)
```

```python
unemploy_2016_df = unemploy_2016_df.reset_index()
unemploy_2018_df = unemploy_2018_df.reset_index()
```

```python
unemploy_2016_df
```

```python
#Renaming Column
unemploy_df_2016 = unemploy_2016_df.rename(columns={"Rank": "2016 Rank",})
unemploy_df_2016.head()
```

```python
#Renaming Column
unemploy_df_2018 = unemploy_2018_df.rename(columns={"Rank": "2018 Rank",})
unemploy_df_2018.head()
```

```python
combine_unemploy = unemploy_df_2016 [['2016 rate','State','2016 Rank']]
```

```python
final_combine_unemploy = combine_unemploy.join(unemploy_df_2018[["2018 rate", "2018 Rank"]])
```

```python
final_combine_unemploy
```

```python
#merging columns 
unemployment_df = final_combine_unemploy[["State", '2016 rate', '2016 Rank', '2018 rate', '2018 Rank']]
unemployment_df
```

```python
unemployment_df.to_csv('unemploymentcombine.csv', index=False, header=True)
```
